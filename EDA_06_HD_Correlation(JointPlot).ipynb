```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns

df = pd.read_csv("heart.csv")

df.head()
```

# **JointPlot**

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df) #Checking corr btw age & maxhr as per out heatmap.
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'hex') #Checking corr btw age & maxhr as per out heatmap.
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'reg') #Checking corr btw age & maxhr as per out heatmap.
```

```python
sns.jointplot(x = 'Age', y = 'Age', data = df) # Showing 100% correlation btw aga & age.
```

```python
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data = df) #Showing No Correlation btw HD And OP.
```

```python
sns.jointplot(x = 'Cholesterol', y = 'Oldpeak', data = df) #Showing No Correlation btw Cholesterol And OP.
```
