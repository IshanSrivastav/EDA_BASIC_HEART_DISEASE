```python
#(PieCharts are Usefull if we have less number of columns)

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns

df = pd.read_csv("heart.csv")

df.head()
```

```python
sns.displot(df['Age'], kde = True, color = 'blue')
```

#### Distribution Plot For Numeric Colums

## RestingBP Distribution

```python
sns.distplot(df['RestingBP'], kde = True, color = 'blue')
```

## Distribution Of Cholesterol

```python
sns.distplot(df['Cholesterol'], kde = True, color = 'blue')
```

## Distribution Of MaxHR

```python
sns.distplot(df['MaxHR'], kde = True, color = 'blue')
```

## Pie Charts For Categorical

```python
df.head()
```

#### Sex

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f')
```

## ChestPainType

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f')
```
